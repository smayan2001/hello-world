# Feature Scaling

### Why feature scaling?
Real Life Datasets have many features with a wide range of values like for example let’s consider the house price prediction dataset. It will have many features like no. of. bedrooms, square feet area of the house, etc.

As you can guess, the no. of bedrooms will vary between 1 and 5, but the square feet area will range from 500-2000. This is a huge difference in the range of both features.

## Which algorithms requires feature scaling?
Many machine learning algorithms that are using Euclidean distance as a metric to calculate the similarities will fail to give a reasonable recognition to the smaller feature, in this case, the number of bedrooms, which in the real case can turn out to be an actually important metric.

Eg: Linear Regression, Logistic Regression, KNN

## Most commonly used feature scaling techniques
1. (Normalization) : Normalization is the process of scaling data into a range of [0, 1]. It's more useful and common for regression tasks.






2. Standardization : Standardization is the process of scaling data so that they have a mean value of 0 and a standard deviation of 1. It's more useful and common for classification tasks.


    

```python
# Import necessary libraries
from sklearn import datasets # Sklearn
import pandas as pd 
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Normalizer
import seaborn as sns 
```

# standardization

```python
#Load the Dataset
iris = datasets.load_iris()
#convert the dataset into a pandas data frame
df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
#display the head (first 5 rows) of the dataset
df.head()
```

```python
# Feature Scaling using Standard scaler
#Standardize the features
#Create an object of StandardScaler which is present in sklearn.preprocessing
scalar = StandardScaler()
scaled_data = pd.DataFrame(scalar.fit_transform(df)) #scaling the data
scaled_data

# convert dataset= training test
#training scalar.fit(X_train)
#testing==scalar.transform(X_test)

#No scaling : Target variable (Y)
```
